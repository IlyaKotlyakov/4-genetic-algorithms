```python
import numpy as np
import pygad

myList=[]
with open('14.txt') as f:
    for line in f:
        myList.append([float(x) for x in line.split()])
lngth=len(myList)

maxWeight=myList[0][0]
maxSize=myList[0][1]

value=[]
size=[]
weight=[]

for i in range (1,lngth):
    #ценность
    value.append(myList[i][2])
    #размер
    size.append(myList[i][1])
    #вес
    weight.append(myList[i][0])
print(myList,'\n')
```

```python

# Функция для подсчёта ценности набора предметов.
# Если в наборе превышен лимит по весу и/или по объёму, то ценность=0

last_fitness = 0

def fitness_func(solution, solution_idx):

    sum1,sum2,sum3 = 0,0,0
    for i in range(len(solution)):
        sum1 +=solution[i] * size[i]
        sum2 += solution[i] * weight[i]
        sum3 += solution[i] * value[i]
    if sum2 > maxWeight or sum1 > maxSize: fitness = 0 
    else: fitness=sum3
    return fitness  


def callback_generation(ga_instance):
    global last_fitness
    print("Поколение = {generation}".format(generation=ga_instance.generations_completed))
    print("Ценность = {fitness}".format(fitness=ga_instance.best_solution()[1]))
    print("Изменение = {change}\n".format(change=ga_instance.best_solution()[1] - last_fitness))
    last_fitness = ga_instance.best_solution()[1]

# Создаём экземпляр класса GA, в конструкторе которого указываем параметры, которые
# будут использованы для получения конечного результата
ga_instance = pygad.GA(num_generations=50, #Количество поколений
                       num_parents_mating=40, #Количество наборов, выбранных в качестве родителей (40/200=20% популяции)
                       fitness_func=fitness_func, #Функция, определяющая ценность набора из популяции
                       sol_per_pop=200, #Количество наборов в популяции
                       num_genes=len(myList)-1, #Количество генов в наборе ({0; 1} для каждого имеющегося предмета, т.е. указывается берётся он в набор или нет)
                       gene_type=np.int16,
                       init_range_low=0, #Нижняя граница случайных значений в наборе начальной популяции
                       init_range_high=2, #Верхняя граница случайных значений в наборе начальной популяции (не включительно)
                       parent_selection_type="random", # Принцип выбора родителей (случайный)
                       keep_parents=40, #20% популяции останутся родителями
                       crossover_type="single_point", # Тип скрещивания
                       mutation_type="random",  # Тип мутаций
                       mutation_percent_genes=20, # Процент мутирующих генов
                       callback_generation=callback_generation)

# Запуск экземпляра класса, начало генерации поколений
ga_instance.run()

# Информация о лучшем наборе в каждом поколении
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Грузы в лучшем наборе : {solution}".format(solution=solution))
print("Ценнсть лучшего набора = {solution_fitness}".format(solution_fitness=solution_fitness))
print("Индекс лучшего набора в популяции : {solution_idx}".format(solution_idx=solution_idx))

if ga_instance.best_solution_generation != -1:
    print("Лучшая ценность была достигнута на {best_solution_generation} поколении.".format(best_solution_generation=ga_instance.best_solution_generation))
```
